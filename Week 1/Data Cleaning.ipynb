```python
clean_outreach = pd.read_excel("Cleaned datasets\Cleaned_Data_outreach.xlsx")
clean_application = pd.read_excel("Cleaned datasets\clean_application.xlsx")
clean_campaign = pd.read_csv('Cleaned datasets\campaign_cleaned.csv')
```

```python
clean_application.head()
```

Merging of 3 datasets 

```python
merge_data = clean_outreach.merge(clean_application, 
                            left_on='Reference_ID', 
                            right_on='App_ID', 
                            how='left')

final_data = merge_data.merge(clean_campaign,
                              left_on="Campaign_ID",
                              right_on="ID",
                              how="left")
```

```python
merge_data
```

```python
final_data
```

```python
final_data.to_excel('clean_data.xlsx',index=False)
```

```python
# Standardize the join keys in both datasets
clean_outreach['Reference_ID'] = clean_outreach['Reference_ID'].astype(str).str.strip().str.upper()
clean_application['App_ID'] = clean_application['App_ID'].astype(str).str.strip().str.upper()
```

```python
merge_data['Campaign_ID'] = merge_data['Campaign_ID'].astype(str).str.strip().str.upper()
clean_campaign['ID'] = clean_campaign['ID'].astype(str).str.strip().str.upper()
```
